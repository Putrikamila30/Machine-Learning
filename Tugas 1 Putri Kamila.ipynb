```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

# Memuat dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Mengatur model
model = LogisticRegression(max_iter=200, solver='liblinear')

# Mengatur K-Fold Cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Melakukan cross-validation dan menghitung skor
scores = cross_val_score(model, X, y, cv=kfold)

# Menampilkan hasil
print(f'K-Fold Cross-validation Scores: {scores}')
print(f'Mean Score: {scores.mean():.4f}')
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression

# Memuat dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Mengatur model
model = LogisticRegression(max_iter=200, solver='liblinear')

# Mengatur Stratified K-Fold Cross-validation
stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Melakukan cross-validation dan menghitung skor
stratified_scores = cross_val_score(model, X, y, cv=stratified_kfold)

# Menampilkan hasil
print(f'Stratified K-Fold Cross-validation Scores: {stratified_scores}')
print(f'Mean Score: {stratified_scores.mean():.2f}')
```

```python
from sklearn.model_selection import LeaveOneOut, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

# Memuat dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Mengatur model
model = LogisticRegression(max_iter=200, solver='liblinear')

# Mengatur Leave-One-Out Cross-validation (LOOCV)
loocv = LeaveOneOut()

# Melakukan cross-validation dan menghitung skor
loocv_scores = cross_val_score(model, X, y, cv=loocv)

# Menampilkan hasil
print(f'Leave-One-Out Cross-validation Mean Score: {loocv_scores.mean():.4f}')
```

```python
from sklearn.model_selection import RepeatedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

# Memuat dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Mengatur model
model = LogisticRegression(max_iter=200, solver='liblinear')

# Mengatur Repeated K-Fold Cross-validation
repeated_kfold = RepeatedKFold(n_splits=5, n_repeats=10, random_state=42)

# Melakukan cross-validation dan menghitung skor
repeated_scores = cross_val_score(model, X, y, cv=repeated_kfold)

# Menampilkan hasil
print(f'Repeated K-Fold Cross-validation Mean Score: {repeated_scores.mean():.4f}')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.datasets import load_iris

# Memuat dataset Iris
iris = load_iris()
X, y = iris.data, iris.target

# Mengatur model SVM
model = SVC()

# Mengatur grid parameter
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf']
}

# Menginisialisasi Grid Search
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Menjalankan Grid Search
grid_search.fit(X, y)

# Menampilkan parameter terbaik dan skor terbaik
print("Best Parameters:", grid_search.best_params_)
print(f"Best Score: {grid_search.best_score_:.4f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Memuat dataset Iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Menambahkan kolom target dan mengonversi angka menjadi nama spesies
data['species'] = iris.target
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
data['species'] = data['species'].map(species_map)

# Menampilkan 5 baris pertama dataset
print("First 5 Rows of the Dataset:")
print(data.head())

# Menampilkan ringkasan statistik dataset
print("\nSummary Statistics:")
print(data.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot untuk melihat hubungan antar variabel
sns.pairplot(data, hue='species', diag_kind='kde')  # Menggunakan KDE untuk distribusi diagonal
plt.show()

# Heatmap untuk melihat korelasi antar fitur
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Feature Correlations")
plt.show()

# Boxplot untuk distribusi panjang sepal berdasarkan spesies
plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=data)
plt.title("Boxplot of Sepal Length by Species")
plt.show()

# Violin plot untuk distribusi lebar sepal berdasarkan spesies
plt.figure(figsize=(12, 6))
sns.violinplot(x='species', y='sepal width (cm)', data=data, inner="quartile")
plt.title("Violin Plot of Sepal Width by Species")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Memisahkan fitur (X) dan target (y)
X = data.drop(columns=['species'])  # Menghapus kolom 'species' dari fitur
y = data['species']

# Membagi data menjadi training (70%) dan testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Melatih model Logistic Regression
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# Memprediksi pada data uji
y_pred = model.predict(X_test)

# Menampilkan hasil evaluasi
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Import library yang dibutuhkan
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Load dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Menambahkan kolom 'species' dengan label target
data['species'] = iris.target
data['species'] = data['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Menampilkan 5 baris pertama dataset
print("First five rows of dataset:\n", data.head())

# Menampilkan statistik deskriptif
print("\nSummary Statistics:\n", data.describe())

# Mengecek nilai yang hilang dalam dataset
print("\nMissing Values:\n", data.isnull().sum())

# Pairplot untuk melihat distribusi antar variabel
sns.pairplot(data, hue='species')
plt.show()

# Heatmap untuk melihat korelasi antar fitur
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap of Feature Correlations")
plt.show()

# Boxplot untuk membandingkan panjang sepal antar spesies
plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=data)
plt.title("Boxplot of Sepal Length by Species")
plt.show()

# Violin plot untuk membandingkan lebar sepal antar spesies
plt.figure(figsize=(12, 6))
sns.violinplot(x='species', y='sepal width (cm)', data=data)
plt.title("Violin Plot of Sepal Width by Species")
plt.show()

# Memisahkan fitur (X) dan target (y)
X = data.drop(columns=['species'])
y = data['species']

# Membagi data menjadi training (70%) dan testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Melatih model Logistic Regression
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# Memprediksi pada data uji
y_pred = model.predict(X_test)

# Menampilkan Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", conf_matrix)

# Menampilkan Classification Report
class_report = classification_report(y_test, y_pred)
print("\nClassification Report:\n", class_report)
```
